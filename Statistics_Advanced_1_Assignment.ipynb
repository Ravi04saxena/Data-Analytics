### Q1. What is a random variable in probability theory?
#### Sol.
A random variable in probability theory is a function that assigns a real number to each outcome in the sample space of a random experiment. This provides a way to numerically describe or quantify the possible results of random processes such as coin tosses, dice rolls, or other chance events.

A random variable does not represent randomness itself, but rather maps unpredictable outcomes to numbers, allowing us to use mathematical and statistical techniques for further analysis.

* There are two main types:
  * Discrete random variables: Assume only specific, separated values (e.g., the number of heads in five coin tosses).
  * Continuous random variables: Can take any value within a continuous range (e.g., the exact height of a person).

In summary, a random variable serves as a bridge between abstract probability theory and real-world data analysis by numerically encoding the outcomes of random events.

### Q2. What are the types of random variables ?
#### Sol. In probability theory, random variables are mainly classified into two broad types -
##### 1. Discrete Random Variables
* Definition: Take countable possible values (finite or infinite countable).
* Examples:
  * Number of heads in 10 coin tosses → {0, 1, 2, …, 10}.
  * Number of students in a class.
  * Number of emails you get in a day.
* Probability distribution: Represented using a Probability Mass Function (PMF), which assigns probabilities to each possible value.

##### 2. Continuous Random Variables
* Definition: Take uncountably infinite possible values in a given interval.
* Examples:
  * Height of a person (can be 170.2 cm, 170.25 cm, etc.).
  * Time it takes to run a marathon.
  * Temperature in a city at noon.
* Probability distribution: Represented using a Probability Density Function (PDF), and probabilities are found over intervals.

### Q3. Explain the difference between discrete and continuous distributions.
#### Sol. Discrete and continuous distributions are two fundamental ways in probability theory to describe how random variables are spread across possible values.

##### Discrete Distributions:
* Definition: Describe the probabilities of outcomes for discrete random variables, which can only take specific, separated values (usually counts or integers).
* Example: The number of heads in five coin tosses, the roll of a die.
* Representation: Probability is assigned to each possible value using a probability mass function (PMF).
* Key property: There are gaps between possible outcomes; values are enumerated (e.g., 0, 1, 2,...).

##### Continuous Distributions:
* Definition: Describe the probabilities for continuous random variables, which can take any value within a given range or interval.
* Example: The exact height of a person, time until a light bulb burns out.
* Representation: Probabilities are expressed using a probability density function (PDF); probabilities correspond to ranges, not exact points.
* Key property: The possible values form a continuum with no gaps; there are infinitely many values within any interval.

### Q4. What is a binomial distribution, and how is it used in probability?
#### Sol. A binomial distribution is a type of discrete probability distribution that describes the number of successes in a fixed number (n) of independent trials, where each trial has only two possible outcomes—commonly labeled as "success" and "failure". Each trial is identical and the probability of success (p) remains the same for every trial.

##### Key Features
* Discrete distribution: Counts the number of successes in events with only two outcomes.
* Bernoulli trials: Each individual trial is called a Bernoulli trial—a random experiment with exactly two possible outcomes.
* Parameters: Defined by the number of trials (n) and probability of success (p) per trial.


image.png

##### Binomial distribution is widely used to model scenarios like:
* Flipping a coin a set number of times (success = heads)
* Measuring the number of defective products in a batch
* Counting survey responses for a yes/no question
* Predicting the number of passes or fails on a test

##### To use the binomial distribution, the process must meet these conditions:
* Fixed number of trials (n)
* Each trial independent of others
* Just two outcomes per trial (often "success" and "failure")
* The probability of success is constant (p)

### Q6. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
#### Sol. The Central Limit Theorem (CLT) is a fundamental principle in statistics. It states that, given a sufficiently large sample size from any population with a finite mean and variance, the sampling distribution of the sample mean will approach a normal (bell-shaped) distribution—even if the original population itself is not normally distributed. Essentially, as you repeatedly sample and calculate their means, these means themselves form a distribution that becomes increasingly normal as sample size grows, typically n≥30 is considered sufficient.

Formal Statement
Let X1, X2.....Xn be independent and identically distributed random variables with mean μ and standard deviation σ. The sample mean "X bar" will be approximately normally distributed:

##### Why is the CLT Critical in Statistics?
* Enables Statistical Inference: CLT underpins much of inferential statistics because it allows analysts to use the normal distribution as an approximation for the distribution of sample means, even when the population distribution is unknown or skewed.
* Simplifies Analysis: By justifying the use of normal-based methods (confidence intervals, hypothesis tests), CLT makes analysis with complex, real-world data both tractable and accurate.
* Universal Applicability: Its power lies in the fact that it applies to nearly all distributions (with finite mean and variance), making statistical methods broadly useful.
* Foundation for Many Tests: Parametric tests (like z-test, t-test) rely on normality—for large samples, CLT ensures sample means will be normal, granting these tests validity in practice.
* Practical Examples:
  * Election polling, quality control, finance, biology, and engineering: CLT allows reliable estimation from sample data.
  * Predicting the average returns on stock portfolios, averaging measurement errors, estimating population parameters from sample survey data.

##### In summary:
The CLT is central to statistics because it assures that the mean of large samples will be normally distributed—even if the population distribution is not—making it possible to apply powerful and flexible statistical techniques for analysis and inference.

### Q7. What is the significance of confidence intervals in statistical analysis?
#### Sol. The significance of confidence intervals in statistical analysis lies in their ability to provide a range of values within which the true population parameter is likely to fall, along with a level of confidence (e.g., 95%). Unlike a single point estimate, confidence intervals communicate the uncertainty around that estimate, helping quantify the precision and reliability of statistical results.

##### Key points about the significance of confidence intervals are:
* Quantify Uncertainty: Confidence intervals describe the variation around a sample estimate and predict the likely range where the true parameter lies if the experiment or sampling is repeated.
* Interpretation of Confidence Levels: A 95% confidence interval means that, under repeated sampling, about 95% of those intervals would contain the true population parameter, providing a long-run measure of reliability for the estimate.
* Assess Statistical Significance: They allow you to see if the interval includes values of no effect (e.g., zero difference). If the confidence interval excludes these null values, this usually indicates statistical significance at the corresponding level.
* Practical Significance and Precision: The width of the confidence interval reflects the estimate's precision. Narrower intervals mean more precise estimates and greater confidence in the result's practical importance.
* Aid Decision Making: By providing a margin of error along with the point estimate, confidence intervals help avoid overconfidence in a single estimate and guide more nuanced conclusions in hypothesis testing, A/B testing, quality control, and many other applications.

In summary, confidence intervals are critical because they communicate the reliability and range of plausible values for estimates, help distinguish significant findings from noise, and enable informed, evidence-based decisions in statistical analysis and applied research.

### Q9. Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.
#### Sol.

```python
import numpy as np
import matplotlib.pyplot as plt

# generate 1000 random numbers from Normal Distribution
mean = 50
std_dev = 5
n = 1000

data = np.random.normal(loc=mean, scale=std_dev, size=n)

# claculate mean and standard deviation
calculated_mean = np.mean(data)
calculated_std_dev = np.std(data)

print(f"Calculated Mean: {calculated_mean:.2f}")
print(f"Calculated Standard Deviation: {calculated_std_dev:.2f}")

# draw histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Draw vertical lines for mean
plt.axvline(calculated_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {calculated_mean:.2f}")
plt.legend()

plt.show()
```

### Q10. You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

* Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
* Write the Python code to compute the mean sales and its confidence interval.

##### Sol.The Central Limit Theorem says that if we repeatedly take random samples from a population and compute their means, the distribution of those means will approach a normal distribution — even if the original data is not normal — provided the sample size is large enough.

##### Here's how it applies to our sales data:
* Population: Daily sales over the 2 years (we have only a subset of 20 days).
* Sample: Our given daily_sales list is a sample of the population.
* Goal: Estimate the population mean (true average daily sales) with 95% confidence.
* Since our sample size is small (n=20), we'll use the t-distribution (because population standard deviation is unknown).
* Formula for 95% Confidence Interval:

image.png

```python
import numpy as np
from scipy import stats

# Sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# claculate mean and standard deviation
mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)  # Sample std deviation
n = len(daily_sales)

# t-critical value for 95% CI
confidence = 0.95
t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

# Margin of Error
margin_of_error = t_critical * (std_sales / np.sqrt(n))

# Confidence Interval
lower_bound = mean_sales - margin_of_error
upper_bound = mean_sales + margin_of_error

print(f"Mean Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```
