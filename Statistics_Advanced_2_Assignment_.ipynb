### Q1. What is hypothesis testing in statistics?
#### Sol. Hypothesis testing is a statistical method used to determine whether there is enough evidence in a sample of data to infer that a certain condition is true for the entire population. It allows analysts and researchers to test assumptions (hypotheses) about population parameters based on sample data, and to draw conclusions about the population as a whole.

#### Key Concepts:
 * Null Hypothesis (H₀): The default or starting assumption that there is no effect or no difference. It often states that a population parameter equals a certain value.

 * Alternative Hypothesis (H₁ or Ha): The opposing assertion, indicating that there is an effect or a difference. This is what the researcher hopes to support.

#### The Hypothesis Testing Process:
The process usually follows these main steps:
State the hypotheses: Define the null and alternative hypotheses.

1. Choose the significance level (α): This is the maximum probability of rejecting the null hypothesis when it is actually true (commonly set at 0.05).

2. Select the appropriate test: Based on the kind of data (e.g., t-test, chi-square, ANOVA).

3. Collect and analyze data: Obtain sample data and compute the test statistic.

4. Calculate the p-value: The probability of obtaining a result at least as extreme as the one observed, assuming the null hypothesis is true.

5. Make a decision:

 * If p-value ≤ α, reject the null hypothesis (suggesting evidence for the alternative hypothesis).

 * If p-value > α, do not reject the null hypothesis (evidence insufficient for the alternative).

7. Interpret and report results: Conclude whether the sample provides enough evidence to support the alternative hypothesis, or if the null hypothesis remains plausible.

### Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
#### Sol. Null hypothesis is the statement of no effect or difference, while the alternative hypothesis is the statement indicating an effect or difference that you want to provide evidence for.

##### Null Hypothesis (H₀):
* The null hypothesis is a default statement asserting that there is no effect, no difference, or no relationship between variables.

* It typically suggests that any observed differences in sample data are due to random chance and not a true effect.

* Example: null hypothesis: The average website visits per day is 50.

##### Alternative Hypothesis (H₁ or Ha):
* The alternative hypothesis is the statement researchers aim to support. It claims that an effect, difference, or relationship does exist.

* It proposes that the population parameter is different from what is stated in the null hypothesis.

* Example: alternative hypothesis: The average website visits per day is not 50.

##### Key Differences:
* Assertion: Null hypothesis(H₀) asserts no change or difference; alternative hypothesis(Ha) asserts the presence of a change or difference.

* Testing Process: Hypothesis testing starts by assuming null hypothesis(H₀) is true. Statistical evidence from the sample is then evaluated to decide whether to reject null hypothesis(H₀) in favor of alternative hypothesis(Ha).

* Decision: If sample evidence is strong (often determined via the p-value), null hypothesis(H₀) may be rejected, suggesting support for alternative hypothesis(Ha). If evidence is weak, null hypothesis(H₀) is not rejected, and no conclusion is drawn about alternative hypothesis(Ha).


### Q3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
#### Sol. The significance level—denoted by the Greek letter α (alpha)—is a key concept in hypothesis testing. It represents the threshold for deciding when to reject the null hypothesis (H₀) in favor of the alternative hypothesis (H₁). Commonly used levels are 0.05 (5%) or 0.01 (1%), but researchers may choose others depending on the context.

#### What Does the Significance Level Mean?
*  The significance level sets the maximum probability of a Type I error, which is the risk of rejecting a true null hypothesis (a "false positive").

* For example, a significance level of α = 0.05 means you're willing to accept a 5% chance that the result you're seeing is actually due to random chance—even if you think it's a real effect.

* This threshold is chosen before collecting data and determines how strong the evidence must be before you reject the null hypothesis.

#### Role in Deciding the Outcome
* After performing a statistical test on your sample data, you calculate a p-value, which tells you how likely your observed result would be under the null hypothesis.

* Decision Rule:

  * If the p-value ≤ α, there is enough evidence to reject the null hypothesis. The result is said to be statistically significant.

  * If the p-value > α, you do not reject the null hypothesis. This means you don't have sufficient evidence to say there's a real effect—your data could simply be due to chance.

#### Why Is It Important?
* The significance level helps balance the risk of making errors while trying to find meaningful results.

* It lets researchers control the likelihood of drawing incorrect conclusions (false positives) from data, and provides a standard for what counts as "strong enough" evidence in statistical tests.

The significance level is the pre-set boundary for deciding whether the evidence from your data is strong enough to reject the null hypothesis. It directly impacts the outcome of hypothesis tests by defining what you consider statistically significant, and thus helps you make more reliable decisions based on sample data.

### Q4. What are Type I and Type II errors? Give examples of each.
#### Sol. In hypothesis testing, two possible errors can occur when making decisions based on sample data:

#### Type I Error (False Positive)
* Definition: Occurs when you reject the null hypothesis (H₀) when it is actually true. In other words, you conclude that there is an effect or difference when there is none.

* Significance Level: The probability of committing a Type I error is called alpha (α), usually set at 0.05.

* Example:
  * Medical Testing: A new diagnostic test identifies healthy individuals as sick (the test says a person has a disease when they actually do not).
  * Legal System: Convicting an innocent person (jury rejects H₀: "Defendant is innocent," when in reality the person is innocent).
  * Quality Control: Rejecting a product as defective that actually meets quality standards.

#### Type II Error (False Negative)
* Definition: Occurs when you fail to reject the null hypothesis when it is actually false. This means you overlook a real effect or difference.

* Probability: The risk of a Type II error is denoted by beta (β).

* Example:

  * Medical Testing: A diagnostic test fails to detect a disease in a sick person (the test says a person is healthy when they are actually sick).

  * Legal System: Acquitting a guilty person (jury fails to reject H₀: "Defendant is innocent," when in reality the person is guilty).

  * Quality Control: Accepting a defective product as meeting quality standards (the product should have been rejected, but was accepted)

### Q5. What is the difference between a Z-test and a T-test? Explain when to use each.
#### Sol.
#### Z-Test
* Used When:
  * The sample size is large (typically ≥30).
  * The population standard deviation is known.
  * The data is approximately normally distributed.
* Distribution Used: Standard normal distribution (Z-distribution).
* Typical Applications: Comparing sample mean to population mean, or means of two large independent samples when the population variance is known.

#### T-Test
* Used When:
  * The sample size is small (typically <30).
  * The population standard deviation is unknown and must be estimated from the sample.
  * The data is approximately normally distributed.
* Distribution Used: T-distribution (which has wider tails than the Z-distribution, accounting for the extra variability in small samples).
* Typical Applications: Comparing sample mean to population mean, or means of two independent groups when the population variance is unknown; also used for paired or dependent samples.

#### How to Choose:
* Use a Z-test when you have a large sample size and know the population standard deviation.
* Use a T-test when your sample is small and the population standard deviation is unknown—you estimate it from your sample.
* If sample size is large and the population standard deviation is unknown, the T-test can still be used, but the Z-test is more powerful if the deviation is known.

#### Example Scenarios
* Z-Test Example: Testing if the average height of 100 randomly selected adult males matches the known population average (with known population standard deviation).
* T-Test Example: Determining whether the average blood pressure reduction from a drug in 15 patients is different from zero (population SD unknown).

### Q6. Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

Hint: Generate random number using random function.
#### Sol.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 10  # number of trials
p = 0.5  # probability of success in each trial
sample_size = 1000  # number of random samples to generate

# Generate random numbers from binomial distribution
samples = np.random.binomial(n, p, sample_size)

# Plot the histogram of the binomial distribution
plt.hist(samples, bins=np.arange(n+2)-0.5, edgecolor='black', density=True)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution Histogram (n=10, p=0.5)')
plt.grid(axis='y', linestyle='--')
plt.show()
```

### Q7.  Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

#### Sol.


```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Null hypothesis population mean (H0)
population_mean = 50

# Significance level
alpha = 0.05

# Compute sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # sample standard deviation
sample_size = len(sample_data)
population_std = 0.5 # (we need population_std for Z-test; if not given, we’ll assume it’s known for this example)

# Calculate Z-statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Critical value for two-tailed test
z_critical = norm.ppf(1 - alpha/2)

# Output
print(f"Sample mean: {sample_mean:.4f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.6f}")
print(f"Critical Z-value: ±{z_critical:.4f}")

# Decision
if abs(z_stat) > z_critical:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

#### Interpretation
* The sample mean (50.0889) is very close to the hypothesized mean of 50.
* The Z-statistic (1.0667) means the difference is only about 1.07 standard errors away from the mean — far less than the ±1.96 cutoff for 95% confidence.
* The p-value (0.2861) is much greater than 0.05, meaning the observed difference could easily occur by random chance if the true mean were 50.
* Conclusion: We fail to reject Null Hypothesis — there’s no statistical evidence that the true mean differs from 50.



### Q8. Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
#### Sol.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate data
np.random.seed(42)  # lock the random generator’s starting point and this make result reproducibility
mu = 50             # true mean
sigma = 5           # true standard deviation
n = 100             # sample size

data = np.random.normal(mu, sigma, n)  # Simulate data from the normal distribution

# Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
standard_error = sample_std / np.sqrt(n)

# Compute 95% confidence interval (using t-distribution)
alpha = 0.05
degree_of_freedom = n-1
t_critical = stats.t.ppf(1 - alpha/2, degree_of_freedom)
margin_of_error = t_critical * standard_error
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample mean: {sample_mean:.3f}")
print(f"Sample Standard Deviation: {sample_std:.3f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Plot data
plt.figure(figsize=(8,5))
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='--', label=f"Lower CI = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='--', label=f"Upper CI = {ci_upper:.2f}")
plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

Explanation:
* The script generates 100 random samples from a normal distribution with mean 50 and standard deviation 5.

* It calculates the sample mean and sample standard deviation.

* The 95% confidence interval for the mean is calculated using the t-distribution (appropriate since the population standard deviation is unknown and sample size is finite).

* The histogram of the data is plotted with vertical lines indicating the sample mean and the lower and upper bounds of the confidence interval.

### Q9. Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
#### Sol.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    data = np.array(data)
    # Calculate the mean and standard deviation of the data
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # sample standard deviation

    # Calculate Z-scores for each data point
    z_scores = (data - mean) / std_dev

    # Plot histogram of Z-scores
    plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7, color='skyblue')
    plt.title('Histogram of Z-scores (Standardized Data)')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.legend()
    plt.show()

    return z_scores

# Example usage
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

z_scores = calculate_and_plot_z_scores(sample_data)
print("Z-scores:", np.round(z_scores, 2))
```
