## 1.What is the difference between a function and a method in Python?
#### Sol. Functions and Method are similar in that both are blocks of reusable code used to perform specific tasks. However, the key difference lies in how they are called and what they are associated with.
* Function - A function is a piece of code defined using the def keyword (or lambda for anonymous functions) that is not associated with any object. It can be called independently.
* Method - A method is a function that is associated with an object (usually defined in a class) and is called using dot (.) notation. It typically takes self as its first parameter to refer to the instance of the class.

### Key Differences -
* Function defined using	"def" keyword	while method define using "def" keyword inside a class.
* Function called independently while method can be called an object or class inrance.
* Function takes usee defined arguments while method takes argument "self" refer to class instance.

```python
# Example Function
def greet(name):
    return "Hello, " + name

print(greet("Ravi"))     # greet is a function and it takes an argument and returns a greeting. It is not tied to any object or class.
```

```python
# Example Method
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):  # This is a method
        return "Hello, " + self.name

p = Person("Ravi")
print(p.greet())        # greet here is a method of the Person class. It is tied to an instance of Person (p).
```

## 6. Explain the concept of generators in Python and how they are defined.
### Ans: Generators are a special type of iterator in Python used to generate values one at a time and using less memory.

#### Key Features of Generators:
* Defined using def and yield (not return)
* Do not store all values in memory
* Produce values lazily (only when needed)
* Automatically implement __iter__() and __next__()

#### How to Define a Generator
* Use the 'yield' keyword inside a function instead of 'return'.



```python
# Example of Generator
def emp_name(*args):
  for name in args:
    yield name           # use 'yield' to create generator

# create generator
name = emp_name('Ravi', 'Rahul', 'Joy', 'Manish')

# Use next() to get names one by one
print(next(name))
print(next(name))
print(next(name))
print(next(name))
```

## 7. What are the advantages of using generators over regular functions?
### Ans:Generators provide several benefits, especially when working with large datasets or streams of data.

#### 1. Memory Efficiency
* Generators do not store all values in memory. They generate each value one by one.
* Regular functions often return full lists, which can consume a lot of memory.

#### 2. Lazy Evaluation
* Generators produce values only when needed, which is ideal for large or infinite sequences.
*  Regular functions compute everything at once.

#### 3. Better Performance in Loops
* Since values are yielded one at a time, there's less overhead during iteration.
* Useful for for loops, pipelines, and streaming data.

#### 4. Simplified Code for Iterators
* Generators automatically implement the iterator protocol (__iter__() and __next__()).
* No need to manually track state or counters.

```python
# Example generator function
def get_numbers():
  for i in range(100000):
    yield i     # generator
gen = get_numbers()
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
# Example Normal  Function
def get_numbers_list():
    return [i for i in range(10)]  # Full list in memory

print(get_numbers_list())
```

## 8.What is a lambda function in Python and when is it typically used?
### Ans:- A lambda function is a small, anonymous function defined using the lambda keyword instead of def. It is used to create simple, one-line functions without a name.

#### Lambda functions are mainly used when you need a short, throwaway functionâ€”usually as an argument to higher-order functions like:

* With "map()" function - apply a function to all items in a list.
* With "filter()" function - filter items based on a condition.
* With sorted() - custom sorting key.
* Inline simple operations.


```python
# Example map()
nums = [2,3,4,5]
square_nums = list(map(lambda x:x**2, nums))
print(square_nums)
```

```python
# Example filter()
lst = [1,2,3,4,5,6]
even_nums = list(filter(lambda x:x%2==0, lst))
print(even_nums)
```

```python
# Example sorted
lst1 = ['Ravi', 'Rahul', 'Manish', 'Joy', 'Medhansh']
sorted_name = list(sorted(lst1, key = lambda x:len(x), reverse = True))
print(sorted_name)
```

```python
# Example Inline simple operation
print((lambda x, y: x+y)(4,5))
```

## 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
### Ans:
* map() function - Applies a function to each element in an iterable and returns the transformed result. Syntax - map(function, iterable)

* reduce() function - Applies a function cumulatively to reduce the iterable to a single value. reduce() function is available in functools module. Syntax - reduce(function, iterable)

* filter() function - Applies a function to each element in an iterable and select elements based on condition from an iterables and returned filtered list. Syntax - filter(function, iterable)


```python
# Example - map()
nums = [1,2,3,4,5,6,7]
cube = map(lambda x:x**3, nums)
print(list(cube))
```

```python
# Example reduce
from functools import reduce
nums = [22,5,77,88,55,34]
total = reduce(lambda x,y:x+y, nums)
print(total)
```

```python
# Example filter
lst = [23,68,34,89,76,12,36, 35]
even = filter(lambda x:x%2==0, lst)
print(list(even))
```

## 1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def even_sum(num):
  evensum = 0
  for i in num:
    if i%2==0:
      evensum = evensum + i
  return evensum

list = [1,2,4,5,6,7,8,9]
result = even_sum(list)
print("Sum of Even numbers : ", result)
```

## 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:
        return False  # 0 and 1 are not prime
    for i in range(2, n):
        if n % i == 0:
            return False  # Not prime if divisible by any number other than 1 and itself
    return True  # Prime if no divisors found

for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is a prime number")
```

# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power_of_two(n):
  for i in range(n+1):
    yield 2**i

for j in power_of_two(4):
  print(j)
```
